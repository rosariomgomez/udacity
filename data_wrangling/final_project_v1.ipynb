### 3.1 Top 10 user contributors

```python
#1. Group by unique user identifier
#2. Add the user name to the pipeline document (to show it in the graph instead of the uid)
#3. Count the number of entries by each user
#4. Sort descending
pipeline = [
            { "$group": { "_id": "$created.uid",
                          "name": { "$first": "$created.user"},
                          "count": { "$sum": 1 } } },
            { "$sort": { "count" : -1 } }
           ]
users = db.bayarea.aggregate(pipeline)['result']
```

```python
#helper methods for plotting
import numpy as np
from matplotlib import pyplot as plt

'''Add text to each graph bars
   Input: bars, subplot object, boolean for correct displaying float values
   Output: subplot object
'''
def autolabel(rects, ax, round_height):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        if round_height:
            height = round(height, 2)
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, height,
                ha='center', va='bottom')
    return ax

'''Plot a bar chart
   Input: x data, y data, figure number, title, xlabel, ylabel, color, figure size, round value (true if data to display is float)
   Output: bar chart plot
'''
def plot_bar_chart(x_data, y_data, num_fig, title='', xlabel='', ylabel='', color='blue', figsize=(8,5), round_values=False):
    fig = plt.figure(num_fig, figsize=figsize)
    ax = plt.subplot(111)
    width = 0.8
    barlist = ax.bar(range(len(y_data)), y_data, color=color)
    ax = autolabel(barlist, ax, round_values)
    ax.set_xticks(np.arange(len(x_data)) + width/2)
    ax.set_xticklabels(x_data, rotation=90)
    ax.set_title("       " + title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    return plt.show()
```

```python
top_users = users[:10] #select the top 10 contributors
user_names = [ user['name'].encode('ascii', 'ignore') for user in top_users]
contributions = [ user['count'] for user in top_users]
plot_bar_chart(user_names, contributions, 1, title="#Contributions by top 10 users", xlabel='user names', ylabel="#contributions", 
               color='purple')
```

### 3.2. Top 20 amenity types

```python
# amenity definition from OSM: Covering an assortment of community facilities including toilets, telephones, banks, pharmacies 
# and schools.
#1. Group by amenity type (if the entry doesn't have amenity, it will be compute as 'None' type)
#2. Count the number of occurrences per type
#3. Sort in descendent order
pipeline = [
            { "$group": { "_id": "$amenity",
                          "count": { "$sum": 1 } } },
            { "$sort": { "count" : -1 } }
           ]
amenities = db.bayarea.aggregate(pipeline)['result']
```

```python
amenit = amenities[1:21] #select top 20. 0 entry in array is 'None' (elements without amenity field)
amenity_names = [ amenity['_id'].encode('ascii', 'ignore') for amenity in amenit]
counts = [ amenity['count'] for amenity in amenit]
plot_bar_chart(amenity_names, counts, 2, title="Top 20 amenity types in the Bay Area", xlabel='amenity types', ylabel="#appearances", 
               color='purple', figsize=(15,8))
```

### 3.3. Top 20 restaurant cuisine types

```python
#1. Match all entries with amenity = restaurant
#2. Group them by cuisine key
#3. Count the number of occurrences per type
#4. Sort descending
pipeline = [
            { "$match": { "amenity": "restaurant"} },
            { "$group": { "_id": "$cuisine" ,
                           "count": { "$sum": 1 } } },
            { "$sort": { "count" : -1 } }
           ]
cuisines = db.bayarea.aggregate(pipeline)['result']
```

```python
cuisin = cuisines[1:21] #select top20.  0 entry is 'None' (elements without cuisine field)
cuisine_names = [ cuisine['_id'].encode('ascii', 'ignore') for cuisine in cuisin]
counts = [ cuisine['count'] for cuisine in cuisin]
plot_bar_chart(cuisine_names, counts, 3, title="Top 20 restaurant cuisine types in the Bay Area", xlabel='cuisine types', ylabel="#appearances", 
               color='purple', figsize=(15,8))
```

### 3.4. Top 20 shop types

```python
#1. Group by shop key
#2. Count
#3. Sort descending
pipeline = [
            { "$group": { "_id": "$shop",
                          "count": { "$sum": 1 } } },
            { "$sort": { "count" : -1 } }
           ]
shops = db.bayarea.aggregate(pipeline)['result']
```

```python
top_shops = shops[1:21] #0 entry is 'None' (elements without shop field)
shop_names = [ shop['_id'].encode('ascii', 'ignore') for shop in top_shops]
counts = [ shop['count'] for shop in top_shops]
plot_bar_chart(shop_names, counts, 3, title="Top 20 shop types in the Bay Area", xlabel='shop types', ylabel="#appearances", 
               color='purple', figsize=(15,8))
```

__There is a shop type called "yes" with 96 entries, which doesn't seem very legit. Let's inspect those entries:__

```python
wrong_shop_names = db.bayarea.find({ "shop": "yes" })
#do they have the amenity tag?
amenities = set()
for shop in wrong_shop_names:
    if 'amenity' in shop.keys():
        amenities.add(shop['amenity'])
amenities
```

```python
#let's fix those ones that include amenity tag, because they don't need to have the shop tag
db.bayarea.update( { "shop" : "yes", "amenity": { "$in" : [ "bar", "fast_food", "fuel", "restaurant" ] } },
                   { "$unset" : { "shop": "" } },
                     multi=True )
```

```python
#check that the unset worked properly
db.bayarea.find( { "shop" : "yes", "amenity": { "$in" : [ "bar", "fast_food", "fuel", "restaurant" ] } }).count()
```

### 3.5. The fashionistas contributors

```python
#look for the top 20 users with more contributions on shop types related to "fashion"
#1. Match the shop type is in the list
#2. Group by user id
#3. Count the number of contributions per each user
#4. Sort in descending order
pipeline = [
            { "$match": { "shop": { "$in" : ["clothes", "beauty", "shoes", "jewelry", "hairdresser" ] } } },
            { "$group": { "_id": "$created.uid",
                          "name": { "$first": "$created.user"},
                          "count": { "$sum": 1 } } },
            { "$sort": { "count" : -1 } }
           ]
fashionistas = db.bayarea.aggregate(pipeline)['result']
```

```python
top_fashionistas = fashionistas[:20]
user_names = [ user['name'].encode('ascii', 'ignore') for user in top_fashionistas]
counts = [ user['count'] for user in top_fashionistas]
plot_bar_chart(user_names, counts, 4, title="Top 20 fashionistas in the Bay Area", xlabel='users', ylabel="#contributions", 
               color='purple', figsize=(15,8))
```

### 3.7. Clothing stores by postal code

```python
#1. Match clothes type shops
#2. Group by postal code
#3. Count them
#4. Sort in descending order
pipeline = [
            { "$match": { "shop": "clothes" } },
            { "$group": { "_id": "$address.postcode",
                          "city": { "$first": "$address.city" },
                          "count": { "$sum": 1 } } },
            { "$sort": { "count" : -1 } }
           ]
clothes_by_pc = db.bayarea.aggregate(pipeline)['result']
```

```python
clothes_pc = clothes_by_pc[1:] #0 entry is 'None' 
pc = [ postal['_id'].encode('ascii', 'ignore') for postal in clothes_pc ]
counts = [ postal['count'] for postal in clothes_pc ]
plot_bar_chart(pc , counts, 5, title="#Clothing stores by Postal Code", xlabel='postal code', ylabel="#clothing stores", 
               color='purple')
```

### 3.8. Top 10 near shops from a given location within 1 km

```python
#1. Create an index on the array pos field
from pymongo import GEO2D
db.bayarea.ensure_index([('pos', GEO2D)])
```

__2.Query the bayarea collection using the "\$near" operator (the query sorts the documents from nearest to farthest)<br>__
  Postal code: 94040 --> Latitude:37.385 Longitude:-122.087<br>
While trying to perform the query, I did: {$near: [37.385, -122.087], $maxDistance: 1000} and it failed. <br>
It seems a MongoDB bug:<br>
http://stackoverflow.com/questions/10020755/mongomapper-near-with-maxdistance-mongooperationfailure-geo-values-have-to<br>
https://jira.mongodb.org/browse/SERVER-1990<br>

```python
near_shops = db.bayarea.find({"pos": {"$maxDistance": 1000, "$near": [37.385, -122.087]}, "shop": {"$exists": 1} }).limit(10)
```

```python
for count, shop in enumerate(near_shops):
    if 'name' in shop.keys():
        print str(count + 1) + '.' + ' Shop category: ' + shop['shop'] + ', name: ' + shop['name']
    else:
        print str(count + 1) + '.' + ' Shop category: ' + shop['shop'] + ' (no name supplied)'
```
